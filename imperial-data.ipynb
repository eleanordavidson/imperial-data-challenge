```python
import numpy as np
import scipy as scp
import pandas as pd
import matplotlib as plt
import seaborn
```

```python
#all_filenames = [x for x in ./stock_data.zip().listdir() if x.endswith('.txt') and os.path.getsize(x) > 0]
all_filenames = ['./a.us.txt', './atnm.us.txt', './hgh.us.txt', './icad.us.txt', './lare.us.txt', 
                 './nurow.us.txt', './rtix.us.txt', './scacu.us.txt', './zts.us.txt', './zyme.us.txt']
filenames = np.random.choice(all_filenames,10)
print(filenames)
```

```python
data = []
for filename in filenames:
    df = pd.read_csv(filename, sep=',')

    label, _, _ = filename.split(sep='.')[1:]
    df['Label'] = label[1:]
    df['Date'] = pd.to_datetime(df['Date'])

    data.append(df)
```

```python
plt.pyplot.figure(figsize=(30, 20))
for i, df in enumerate(data):
    axis = plt.pyplot.subplot(5, 2, i+1)
    close_prices = df[['Date', 'Close']].set_index('Date')
    market = df[['Label']].iloc[0][0]
    close_prices.plot(ax=axis, title=market, legend=False, fontsize=16)
plt.pyplot.suptitle('Closing Prices', fontsize=20)
```
