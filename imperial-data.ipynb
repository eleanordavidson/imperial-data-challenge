```python
import numpy as np
import scipy as scp
import pandas as pd
import matplotlib as plt
import seaborn
```

```python
filenames = ['./data/a.us.txt', './data/atnm.us.txt', './data/hgh.us.txt', './data/icad.us.txt', './data/lare.us.txt', 
                 './data/nurow.us.txt', './data/rtix.us.txt', './data/scacu.us.txt', './data/zts.us.txt', './data/zyme.us.txt']
                 
data = []
for filename in filenames:
    df = pd.read_csv(filename, sep=',')

    label, _, _ = filename.split('.')[1:]
    df['Label'] = label[6:]
    df['Date'] = pd.to_datetime(df['Date'])

    data.append(df)
```

```python
plt.pyplot.figure(figsize=(25, 25))
for i, df in enumerate(data):
    axis = plt.pyplot.subplot(5, 2, i+1)
    close_prices = df[['Date', 'Close']].set_index('Date')
    market = df[['Label']].iloc[0][0]
    close_prices.plot(ax=axis, legend=False, fontsize=14)
    axis.set_title(market, fontsize=20)
plt.pyplot.suptitle('Closing Prices', fontsize=30)
plt.pyplot.subplots_adjust(left=None, bottom=None, right=None, top=0.93, wspace=None, hspace=0.5)
```

```python
plt.pyplot.figure(figsize=(25, 25))
for i, df in enumerate(data):
    axis = plt.pyplot.subplot(5, 2, i+1)
    volumes = df[['Date', 'Volume']].set_index('Date')
    market = df[['Label']].iloc[0][0]
    volumes.plot(ax=axis, legend=False, fontsize=14)
    axis.set_title(market, fontsize=20)
plt.pyplot.suptitle('Volume', fontsize=30)
plt.pyplot.subplots_adjust(left=None, bottom=None, right=None, top=0.93, wspace=None, hspace=0.5)
```
